## Aufgabe 2.7
Was unterscheidet den Wert None vom Wert "None"?
None ist eine null variable -> pointer auf null object -> NoneType
"None" wäre eigentlich ein String...in Dataframes ein object

Was den Wert 5 vom Wert "5"?
5 ist int
"5" ist String

Was den Wert "NaN" vom Wert NaN?
"NaN" ist ein String
NaN ist Pandas null variable -> Not a Number -> numpy.float64


Ist True und "True" das gleiche?
Nein, True ist bool, "True" ist String, daher !=

```python
i = None
j = "None"
print(type(i))
print(type(j))
```
